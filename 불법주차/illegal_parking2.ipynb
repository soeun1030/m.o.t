```python
import pandas as pd

# CSV 파일 경로
file_path = "C:/Users/soeun/codingon/project1/불법주차/불법주정차 단속 위치정보(23년11월1일_24년3월13일)_3.csv"

# CSV 파일 로드
data = pd.read_csv(file_path)

# 날짜 필터링을 위해 단속일을 문자열로 처리 후 23년 12월 31일까지의 자료만 추출
data = data[data['단속일'].astype(str) <= '20231231']

# "구" 추출 (구주소에서 두 번째 단어로 추출)
data['구'] = data['구주소'].str.split().str[1]

# "구"별로 단속 건수 계산
region_counts = data['구'].value_counts()

# 결과를 DataFrame으로 변환
region_counts_df = region_counts.reset_index()[:25]
region_counts_df.columns = ['구', '단속건수']

# 결과 출력
print(region_counts_df)

# 필요 시 데이터를 저장
region_counts_df.to_csv('지역구별_단속건수_3.csv', index=False, encoding="utf-8-sig")
```

```python
import pandas as pd

# 각 파일 경로 설정
file1_path = '지역구별_단속건수_1.csv'
file2_path = '지역구별_단속건수_2.csv'
file3_path = '지역구별_단속건수_3.csv'

# 각 파일 로드
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)
df3 = pd.read_csv(file3_path)

# 데이터 병합 (세 DataFrame을 하나로 결합)
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# 지역구별 단속 건수 합산
summed_df = combined_df.groupby('구', as_index=False)['단속건수'].sum()

# 결과를 새로운 CSV 파일로 저장
output_path = '지역구별_단속건수_합계.csv'
summed_df.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"지역구별 단속 건수 합산 결과를 {output_path}에 저장했습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 합계 결과 CSV 파일 경로
file_path = '지역구별_단속건수_합계.csv'

# 데이터 로드
data = pd.read_csv(file_path)
data

# 가나다순으로 정렬
data_sorted = data.sort_values(by='구')

# 바차트 생성
plt.figure(figsize=(12, 7))  # 차트 크기 설정
bars = plt.bar(data_sorted['구'], data_sorted['단속건수'], color='skyblue')
plt.rc('font', family='Malgun Gothic')

# 차트 제목 및 축 레이블 설정
plt.title('지역구별 단속 건수 (가나다순)', fontsize=16)
plt.xlabel('지역구', fontsize=12)
plt.ylabel('단속 건수', fontsize=12)

# x축 레이블 회전 (가독성 향상)
plt.xticks(rotation=45, ha='right')

# 각 바 위에 수치 표시
for bar in bars:
    height = bar.get_height()  # 막대의 높이(값)
    plt.text(bar.get_x() + bar.get_width() / 2, height,  # 텍스트 위치 설정
             f'{int(height):,}',  # 값 표시 (쉼표 추가)
             ha='center', va='bottom', fontsize=10)  # 텍스트 가운데 정렬

# 차트 저장 (옵션)
plt.tight_layout()
plt.savefig('지역구별_단속건수_가나다순_수치포함.png', dpi=300)

# 차트 표시
plt.show()
```

```python
import pandas as pd
import folium
import json

csv_file_path = '지역구별_단속건수_합계.csv'
geojson_file_path = 'Seoul_Gu.json'
population_data = pd.read_csv(csv_file_path, encoding='utf-8')
population_data['단속건수'] = (
    population_data['단속건수']
    .fillna('0')  # Replace NaN with '0'
    .astype(str)  # Convert all values to strings
    .str.replace(',', '')  # Remove commas
    .astype(float)  # Convert to numeric
)


district_population = population_data.groupby('구')['단속건수'].sum().reset_index()
district_population.rename(columns={'구': 'SIG_KOR_NM', '단속건수': 'population_density'}, inplace=True)

with open(geojson_file_path, 'r', encoding='EUC-KR') as file:
    geojson_data = json.load(file)

center = [37.541, 126.986]  # 서울 중심 좌표
m = folium.Map(location=center, zoom_start=11)

#구별 인구
folium.Choropleth(
    geo_data=geojson_data,
    data=district_population,
    columns=['SIG_KOR_NM', 'population_density'],
    key_on='properties.SIG_KOR_NM',
    fill_color='OrRd',
    legend_name='구별 불법주차'
).add_to(m)

m.save('불법주차_map.html')
m
```
