```python
import pandas as pd
import matplotlib.pyplot as plt

#  
time_data = {
    "": ["", "", "", "", "", "", "", "", "", "",
             "", "", "", "", "", "", "", "", "", "",
             "", "", "", "", ""],
    "(s)": [396.83, 332.03, 325.64, 323.46, 379.83, 358.08, 356.34, 393.95, 378.04, 337.38,
                   343.75, 392.67, 302.33, 354.45, 398.09, 340.98, 322.55, 357.51, 351.69, 355.65,
                   331.16, 348.92, 370.60, 326.23, 372.97]
}

violation_data = {
    "": ["", "", "", "", "", "", "", "", "", "",
             "", "", "", "", "", "", "", "", "", "",
             "", "", "", "", ""],
    "": [154976, 58072, 29519, 50900, 56421, 51463, 59698, 39905, 45585, 43351,
               52707, 32377, 117912, 19347, 71225, 30609, 36450, 73354, 55885, 124197,
               65774, 47642, 61002, 131793, 25827]
}

#  
df_time = pd.DataFrame(time_data)
df_violation = pd.DataFrame(violation_data)

#   ( )
merged_data = pd.merge(df_time, df_violation, on="")

#    (  )
merged_data["()"] = merged_data[""] / 10000

#  
fig, ax1 = plt.subplots(figsize=(14, 8))

#   :  
color = 'blue'
ax1.set_xlabel("", fontsize=12)
ax1.set_ylabel("  ()", fontsize=12, color=color)
ax1.plot(merged_data[""], merged_data["(s)"], marker='o', color=color, label="  ()")
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(merged_data[""], rotation=45, ha='right', fontsize=10)

#   :  ( )
ax2 = ax1.twinx()
color = 'orange'
ax2.set_ylabel(" ( )", fontsize=12, color=color)
ax2.plot(merged_data[""], merged_data["()"], marker='s', color=color, label=" ( )")
ax2.tick_params(axis='y', labelcolor=color)

#    
fig.suptitle("     ", fontsize=16)
fig.tight_layout()

#  
ax1.legend(loc="upper left", fontsize=12)
ax2.legend(loc="upper right", fontsize=12)

#  
plt.show()
```
