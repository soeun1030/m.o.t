```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
time_data = {
    "지역구": ["강남구", "강동구", "강북구", "강서구", "관악구", "광진구", "구로구", "금천구", "노원구", "도봉구",
             "동대문구", "동작구", "마포구", "서대문구", "서초구", "성동구", "성북구", "송파구", "양천구", "영등포구",
             "용산구", "은평구", "종로구", "중구", "중랑구"],
    "소요시간(s)": [396.83, 332.03, 325.64, 323.46, 379.83, 358.08, 356.34, 393.95, 378.04, 337.38,
                   343.75, 392.67, 302.33, 354.45, 398.09, 340.98, 322.55, 357.51, 351.69, 355.65,
                   331.16, 348.92, 370.60, 326.23, 372.97]
}

violation_data = {
    "지역구": ["강남구", "강동구", "강북구", "강서구", "관악구", "광진구", "구로구", "금천구", "노원구", "도봉구",
             "동대문구", "동작구", "마포구", "서대문구", "서초구", "성동구", "성북구", "송파구", "양천구", "영등포구",
             "용산구", "은평구", "종로구", "중구", "중랑구"],
    "단속건수": [154976, 58072, 29519, 50900, 56421, 51463, 59698, 39905, 45585, 43351,
               52707, 32377, 117912, 19347, 71225, 30609, 36450, 73354, 55885, 124197,
               65774, 47642, 61002, 131793, 25827]
}

# 데이터프레임 생성
df_time = pd.DataFrame(time_data)
df_violation = pd.DataFrame(violation_data)

# 데이터 병합 (지역구 기준)
merged_data = pd.merge(df_time, df_violation, on="지역구")

# 단속건수 단위를 조정 (만 단위로 나눔)
merged_data["단속건수(만)"] = merged_data["단속건수"] / 10000

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(14, 8))

# 첫 번째 축: 평균 소요시간
color = 'blue'
ax1.set_xlabel("지역구", fontsize=12)
ax1.set_ylabel("평균 소요시간 (초)", fontsize=12, color=color)
ax1.plot(merged_data["지역구"], merged_data["소요시간(s)"], marker='o', color=color, label="평균 소요시간 (초)")
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(merged_data["지역구"], rotation=45, ha='right', fontsize=10)

# 두 번째 축: 단속건수 (만 단위)
ax2 = ax1.twinx()
color = 'orange'
ax2.set_ylabel("단속건수 (만 건)", fontsize=12, color=color)
ax2.plot(merged_data["지역구"], merged_data["단속건수(만)"], marker='s', color=color, label="단속건수 (만 건)")
ax2.tick_params(axis='y', labelcolor=color)

# 제목 및 범례 설정
fig.suptitle("서울특별시 지역구별 평균 소요시간 및 단속건수", fontsize=16)
fig.tight_layout()

# 범례 추가
ax1.legend(loc="upper left", fontsize=12)
ax2.legend(loc="upper right", fontsize=12)

# 그래프 출력
plt.show()
```
