```python
import geopandas as gpd
import matplotlib.pyplot as plt

# GeoJSON    (  )
geojson_path = "C:/Users/soeun/OneDrive/Desktop/ 1-  /seoul_municipalities_geo_simple.json"

# GeoJSON  
seoul_map = gpd.read_file(geojson_path)

#  
fig, ax = plt.subplots(figsize=(10, 10))
seoul_map.plot(ax=ax, edgecolor='black', color='lightblue', alpha=0.7)
ax.set_title("  ", fontsize=15)
ax.axis('off')

plt.show()
```
