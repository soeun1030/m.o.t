```python
import geopandas as gpd
import matplotlib.pyplot as plt

# GeoJSON 파일 경로 설정 (백슬래시를 슬래시로 대체)
geojson_path = "C:/Users/soeun/OneDrive/Desktop/스팩 프로젝트1- 데이터 분석/seoul_municipalities_geo_simple.json"

# GeoJSON 파일 읽기
seoul_map = gpd.read_file(geojson_path)

# 지도 시각화
fig, ax = plt.subplots(figsize=(10, 10))
seoul_map.plot(ax=ax, edgecolor='black', color='lightblue', alpha=0.7)
ax.set_title("서울시 구별 지도", fontsize=15)
ax.axis('off')

plt.show()
```
