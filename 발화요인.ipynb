```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') 

file_path = '출동현황1.csv'
Cause = pd.read_csv(file_path, encoding='utf-8')

value_counts = Cause['발화요인'].value_counts()
print(value_counts)

plt.figure(figsize=(10, 6))
plt.title('발화요인', fontsize=20, weight='bold')
plt.xlabel('', fontsize=15)
plt.ylabel('')

value_counts.plot.barh(color='red', edgecolor='black')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


# for index, value in enumerate(value_counts):
#     plt.text(value + 0.5, index, str(value), fontsize=12, va='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '출동현황.csv'
Cause = pd.read_csv(file_path, encoding='utf-8')
# print(Cause.columns)
value_counts = Cause['발화요인'].value_counts()
print(value_counts)

plt.figure(figsize=(7, 7))
plt.title('발화요인', fontsize=20)  
plt.ylabel('')

value_counts.plot.pie(
    autopct='%.2f%%', colors=plt.cm.Paired.colors,
    textprops={'fontsize': 10, 'weight': 'bold'})

plt.show()  
```
