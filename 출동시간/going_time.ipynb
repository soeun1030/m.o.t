```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# 데이터 로드
data = pd.read_excel("지역구별_평균_소요시간.xlsx")

# 데이터 확인
print(data.head())

# 바 차트 생성
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.rc('font', family='Malgun Gothic') 

# 색상을 조건에 따라 지정
colors = [
    'black' if time > 420 else 'red' if time > 360 else 'orange' if time > 300 else 'blue'
    for time in data['소요시간(s)']
]

# 바 차트 그리기
bars = plt.bar(data['지역구'], data['소요시간(s)'], color=colors)

# y축 눈금 설정 (60 단위)
y_max = data['소요시간(s)'].max()  # 데이터에서 최대값 확인
plt.yticks(np.arange(0, y_max + 60, 60))  # 60 단위로 눈금 설정

# 300초 기준선 추가
# plt.axhline(y=300, color='black', linestyle="--", linewidth=1.5, label="300초 (골든타임)")


# 범례 추가를 위한 가짜 막대 생성
legend_elements = [
    Patch(facecolor='orange', edgecolor='black', label='300~360초'),
    Patch(facecolor='red', edgecolor='black', label='360~420초'),
    Patch(facecolor='black', edgecolor='black', label='420초 이상'),
    plt.axhline(y=300, color='green', linestyle="--", linewidth=1.5, label="300초 (골든타임)")
]

plt.legend(handles=legend_elements, loc='upper right')

# 그래프 제목과 레이블 설정
plt.title('서울특별시 지역구별 평균 출동 소요 시간', fontsize=14)
plt.xlabel('지역구', fontsize=12)
plt.ylabel('평균 출동 소요 시간(s)', fontsize=12)
plt.xticks(rotation=45, ha='right')  # x축 레이블 각도 조정

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 저장 또는 출력
plt.savefig('지역구별_평균소요시간_색상범례바차트.png')  # 그래프를 파일로 저장
plt.show()  # 그래프 출력
```

```python
import pandas as pd
import folium
import json

csv_file_path = '지역구별_평균_소요시간.csv'
geojson_file_path = 'Seoul_Gu.json'
population_data = pd.read_csv(csv_file_path, encoding='utf-8')
population_data['소요시간(s)'] = (
    population_data['소요시간(s)']
    .fillna('0')  # Replace NaN with '0'
    .astype(str)  # Convert all values to strings
    .str.replace(',', '')  # Remove commas
    .astype(float)  # Convert to numeric
)


district_population = population_data.groupby('지역구')['소요시간(s)'].sum().reset_index()
district_population.rename(columns={'지역구': 'SIG_KOR_NM', '소요시간(s)': 'population_density'}, inplace=True)

with open(geojson_file_path, 'r', encoding='EUC-KR') as file:
    geojson_data = json.load(file)

center = [37.541, 126.986]  # 서울 중심 좌표
m = folium.Map(location=center, zoom_start=11)

#구별 인구
folium.Choropleth(
    geo_data=geojson_data,
    data=district_population,
    columns=['SIG_KOR_NM', 'population_density'],
    key_on='properties.SIG_KOR_NM',
    fill_color='OrRd',
    legend_name='구별 출동시간'
).add_to(m)

m.save('출동시간_map.html')
m
```
