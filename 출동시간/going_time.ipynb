```python
import pandas as pd
data = pd.read_excel("지역구별_평균_소요시간.xlsx")
data

import matplotlib.pyplot as plt
import numpy as np

# 데이터 확인
print(data.head())

# 바 차트 생성
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.bar(data['지역구'], data['소요시간(s)'])

# y축 눈금 설정 (60 단위)
y_max = data['소요시간(s)'].max()  # 데이터에서 최대값 확인
plt.yticks(np.arange(0, y_max + 60, 60))  # 60 단위로 눈금 설정

# 300초 기준선 추가
plt.axhline(y=300, color='red', linestyle="--", linewidth=1.5, label="기준선: 300초")

# 그래프 제목과 레이블 설정
plt.title('서울특별시 지역구별 평균 출동 소요 시간', fontsize=14)
plt.xlabel('지역구', fontsize=12)
plt.ylabel('평균 출동 소요 시간(s)', fontsize=12)
plt.xticks(rotation=45, ha='right')  # x축 레이블 각도 조정

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 저장 또는 출력
plt.savefig('지역구별_평균소요시간_바차트.png')  # 그래프를 파일로 저장
plt.show()  # 그래프 출력
```
