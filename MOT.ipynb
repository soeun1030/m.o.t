```python
import matplotlib.pyplot as plt
from adjustText import adjust_text  # 겹침 방지 라이브러리
plt.rc('font', family='Malgun Gothic')

# 데이터프레임에서 필요한 열만 추출하고 소방서명에서 "소방서" 제거
filtered_data = data[['소방서명', '면적 (㎢)', '소방센터 개수']].copy()
filtered_data['소방서명'] = filtered_data['소방서명'].str.replace('소방서', '', regex=False)

# 산점도 생성
plt.figure(figsize=(12, 8))
plt.scatter(filtered_data['면적 (㎢)'], filtered_data['소방센터 개수'], c='blue', alpha=0.7, s=100)

# 각 점에 소방서명을 표시하기 위해 텍스트 객체 생성
texts = []
for i in range(len(filtered_data)):
    texts.append(plt.text(filtered_data['면적 (㎢)'].iloc[i], 
                          filtered_data['소방센터 개수'].iloc[i], 
                          filtered_data['소방서명'].iloc[i],
                          fontsize=10, ha='center', va='bottom', color='darkblue'))

# 텍스트 위치 조정
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

# 축과 제목 설정
plt.xlabel('면적 (㎢)', fontsize=12)
plt.ylabel('소방센터 개수', fontsize=12)
plt.title('면적과 소방센터 개수의 관계', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
incident_counts_file_path = '소방서별_사건_총_횟수.xlsx'
carelessness_file_path = '부주의.xlsx'

# 파일 읽기
incident_counts_data = pd.read_excel(incident_counts_file_path)
carelessness_data = pd.read_excel(carelessness_file_path)

# 사건 총 횟수와 부주의 총 횟수를 병합
correlation_data = pd.merge(
    incident_counts_data[['소방서명', '사건 총 횟수']],
    carelessness_data[['소방서명', '부주의 총 횟수']],
    on='소방서명'
)

# 산점도 생성
plt.figure(figsize=(10, 6))
plt.scatter(correlation_data['사건 총 횟수'], correlation_data['부주의 총 횟수'], c='blue', alpha=0.7, s=100)

# 각 점에 소방서명을 표시
for i in range(len(correlation_data)):
    plt.text(correlation_data['사건 총 횟수'].iloc[i], 
             correlation_data['부주의 총 횟수'].iloc[i] + 5,  # 텍스트를 약간 위로 조정
             correlation_data['소방서명'].iloc[i],
             fontsize=9, ha='center', color='darkblue')

# 그래프 설정
plt.xlabel('사건 총 횟수', fontsize=12)
plt.ylabel('부주의 총 횟수', fontsize=12)
plt.title('사건 총 횟수와 부주의 총 횟수의 상관관계', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# 상관계수 계산
correlation_coefficient = correlation_data[['사건 총 횟수', '부주의 총 횟수']].corr().iloc[0, 1]
print(f"사건 총 횟수와 부주의 총 횟수의 상관계수: {correlation_coefficient:.2f}")
```

```python
import pandas as pd
import folium
import json

# CSV 및 GeoJSON 파일 경로
csv_file_path = '인구밀도2.csv'
geojson_file_path = 'Seoul_Gu.json'

# CSV 파일 로드
population_data = pd.read_csv(csv_file_path, encoding='utf-8')
# 인구 열에서 쉼표 제거 및 숫자로 변환
population_data['인구 (명)'] = population_data['인구 (명)'].str.replace(',', '').astype(float)

# 동별 데이터를 구별로 합산
district_population = population_data.groupby('동별(2)')['인구 (명)'].sum().reset_index()
district_population.rename(columns={'동별(2)': 'SIG_KOR_NM', '인구 (명)': 'population_density'}, inplace=True)

# GeoJSON 파일 로드
with open(geojson_file_path, 'r', encoding='EUC-KR') as file:
    geojson_data = json.load(file)

# 서울 중심 좌표
center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=11)

# 구별 인구수 Choropleth 추가
folium.Choropleth(
    geo_data=geojson_data,
    data=district_population,
    columns=['SIG_KOR_NM', 'population_density'],
    key_on='properties.SIG_KOR_NM',  # GeoJSON 속성 이름과 일치해야 함
    fill_color='PuBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='구별 인구 (명)'
).add_to(m)

# HTML 파일로 저장
m.save('구별_인구수_map.html')

# 맵 객체 반환
m
```

```python
import pandas as pd
import folium
import json
import chardet

# CSV 파일 인코딩 확인 및 로드
csv_file_path = '소방자동차_진입_장애_현황.csv'
with open(csv_file_path, 'rb') as file:
    result = chardet.detect(file.read())
    encoding = result['encoding']

fire_data = pd.read_csv(csv_file_path, encoding=encoding)
fire_data.rename(columns={'소계': 'total_fire_block'}, inplace=True)

# GeoJSON 파일 인코딩 확인 및 로드
geojson_file_path = 'Seoul_Gu.json'
with open(geojson_file_path, 'rb') as file:
    result = chardet.detect(file.read())
    geojson_encoding = result['encoding']

with open(geojson_file_path, 'r', encoding=geojson_encoding) as file:
    geojson_data = json.load(file)

# 서울 지도 생성
center = [37.541, 126.986]  # 서울 중심 좌표
m = folium.Map(location=center, zoom_start=11)

# Choropleth 레이어 추가
folium.Choropleth(
    geo_data=geojson_data,
    data=fire_data,
    columns=['SIG_KOR_NM', 'total_fire_block'],
    key_on='properties.SIG_KOR_NM',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='소방 진입 장애 총 횟수'
).add_to(m)

# 지도 저장 및 출력

m
```

```python
import pandas as pd
import folium
import json
import chardet

# CSV 파일 로드
csv_file_path = '소방자동차_진입_장애_현황.csv'
with open(csv_file_path, 'rb') as file:
    result = chardet.detect(file.read())
    encoding = result['encoding']

fire_data = pd.read_csv(csv_file_path, encoding=encoding)
fire_data.rename(columns={'소계': 'total_fire_block'}, inplace=True)

# GeoJSON 파일 로드
geojson_file_path = 'Seoul_Gu.json'
with open(geojson_file_path, 'rb') as file:
    result = chardet.detect(file.read())
    geojson_encoding = result['encoding']

with open(geojson_file_path, 'r', encoding=geojson_encoding) as file:
    geojson_data = json.load(file)

# 데이터 일치 여부 확인 및 수정
fire_gu_names = set(fire_data['SIG_KOR_NM'])
geojson_gu_names = {feature['properties']['SIG_KOR_NM'] for feature in geojson_data['features']}

# 필요한 경우 이름 수정
fire_data['SIG_KOR_NM'] = fire_data['SIG_KOR_NM'].replace({'서울특별시 종로구': '종로구'})

# 결측치 처리
fire_data['total_fire_block'].fillna(0, inplace=True)

# 서울 지도 생성
center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=11)

# Choropleth 레이어 추가
folium.Choropleth(
    geo_data=geojson_data,
    data=fire_data,
    columns=['SIG_KOR_NM', 'total_fire_block'],
    key_on='properties.SIG_KOR_NM',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    threshold_scale=[0, 10, 20, 30, 40, 50],  # 데이터 범위에 맞게 설정
    legend_name='소방 진입 장애 총 횟수'
).add_to(m)

# 지도 저장 및 출력

m
```

```python
import pandas as pd
import folium
import json

csv_file_path = '소방자동차_진입_장애_현황.csv'
geojson_file_path = 'Seoul_Gu.json'
population_data = pd.read_csv(csv_file_path, encoding='utf-8')
population_data['소계'] = (
    population_data['소계']
    .fillna('0')  # Replace NaN with '0'
    .astype(str)  # Convert all values to strings
    .str.replace(',', '')  # Remove commas
    .astype(float)  # Convert to numeric
)


district_population = population_data.groupby('SIG_KOR_NM')['소계'].sum().reset_index()
district_population.rename(columns={'SIG_KOR_NM': 'SIG_KOR_NM', '소계': 'population_density'}, inplace=True)

with open(geojson_file_path, 'r', encoding='EUC-KR') as file:
    geojson_data = json.load(file)

center = [37.541, 126.986]  # 서울 중심 좌표
m = folium.Map(location=center, zoom_start=11)

#구별 인구
folium.Choropleth(
    geo_data=geojson_data,
    data=district_population,
    columns=['SIG_KOR_NM', 'population_density'],
    key_on='properties.SIG_KOR_NM',
    fill_color='OrRd',
    legend_name='구별 진입장애도'
).add_to(m)

m.save('구별_진입장애도_map.html')
m
```

```python
import pandas as pd

# 업로드된 CSV 파일 경로
csv_file_path = '소방자동차_진입_장애_현황.csv'

# CSV 파일 읽기
data = pd.read_csv(csv_file_path, encoding='utf-8')

# 필요한 열의 데이터 처리
data['소계'] = (
    data['소계']
    .fillna('0')  # NaN을 '0'으로 대체
    .astype(str)  # 문자열로 변환
    .str.replace(',', '')  # 쉼표 제거
    .astype(float)  # 숫자로 변환
)

# 구별로 장애도 합산
district_data = data.groupby('SIG_KOR_NM')['소계'].sum().reset_index()
district_data.rename(columns={'SIG_KOR_NM': '구', '소계': '장애도'}, inplace=True)

# 장애도가 가장 큰 구와 가장 작은 구 찾기
max_district = district_data.loc[district_data['장애도'].idxmax()]
min_district = district_data.loc[district_data['장애도'].idxmin()]

max_district, min_district
```
